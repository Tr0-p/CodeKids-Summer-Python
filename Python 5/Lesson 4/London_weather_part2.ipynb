### Exploring Weather Trends

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the data from the excel file
weather = pd.read_excel('/content/london_weather.xlsx')

# Fill the missing values with mean data
weather_clean = weather.fillna(weather.mean())

# Adjust the data format to date format
weather_clean['date'] = pd.to_datetime(weather_clean['date'], format='%Y%m%d')
```

### Lesson 4

```python
# Filter and generate chart for the last 5 years

# Define the start and end dates
latest_date = weather_clean['date'].max()
start_date = latest_date - pd.DateOffset(years=5)

# Filter the dataset for the last 5 years
last5years = weather_clean[(weather_clean['date'] >= start_date) & (weather_clean['date'] <= latest_date)]
```

```python
# Plotting the maximum and minimum temperatures in London
plt.figure(figsize=(12, 8)) # Set up figure size
plt.plot(last5years['date'], last5years['max_temp'], label='Max Temperature', color='red') # Plot maximum temperature
plt.plot(last5years['date'], last5years['min_temp'], label='Min Temperature', color='blue') # Plot minimum temperature

# Add chart title and labels
plt.title('Daily Maximum and Minimum Temperatures')
plt.xlabel('Date')
plt.ylabel('Temperature (Â°C)')

# Add legend
plt.legend()

# Display the chart
plt.show()
```

```python
# Filter and generate chart for the last 6 months

# Define the start and end dates
latest_date = weather_clean['date'].max()
start_date = latest_date - pd.DateOffset(months=6)

# Filter the dataset for the last 5 years
last6months = weather_clean[(weather_clean['date'] >= start_date) & (weather_clean['date'] <= latest_date)]
```

```python
# Plotting the maximum and minimum temperatures in London
plt.figure(figsize=(12, 8)) # Set up figure size
plt.plot(last6months['date'], last6months['max_temp'], label='Max Temperature', color='red') # Plot maximum temperature
plt.plot(last6months['date'], last6months['min_temp'], label='Min Temperature', color='blue') # Plot minimum temperature

# Add chart title and labels
plt.title('Daily Maximum and Minimum Temperatures')
plt.xlabel('Date')
plt.ylabel('Temperature (Â°C)')

# Add legend
plt.legend()

# Display the chart
plt.show()
```

```python
# Resample daily temperature to monthly temperature
monthlydata = last5years.set_index('date').resample('M').mean()
```

```python
# Plotting the maximum and minimum temperatures in London
plt.figure(figsize=(12, 8)) # Set up figure size
plt.plot(monthlydata.index, monthlydata['max_temp'], label='Max Temperature', marker='o')
plt.plot(monthlydata.index, monthlydata['min_temp'], label='Min Temperature', marker='o')

# Add chart title and labels
plt.title('Seasonal Monthly Temperature Trends')
plt.xlabel('Date')
plt.ylabel('Temperature (Â°C)')

# Add legend
plt.legend()

# Display the chart
plt.show()
```
