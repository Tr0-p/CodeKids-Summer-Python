### Exploring Weather Trends

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the data from the excel file
weather = pd.read_excel('/content/london_weather.xlsx')
print("Data ran successfully:")
weather.head()
```

```python
weather.isnull().sum()
```

```python
# Fill the missing values with mean data
weather_clean = weather.fillna(weather.mean())
```

```python
weather_clean.isnull().sum()
```

```python
# Plotting the maximum and minimum temperatures in London
plt.figure(figsize=(12, 8)) # Set up figure size
plt.plot(weather_clean['date'], weather_clean['max_temp'], label='Max Temperature', color='red') # Plot maximum temperature
plt.plot(weather_clean['date'], weather_clean['min_temp'], label='Min Temperature', color='blue') # Plot minimum temperature

# Add chart title and labels
plt.title('Dailt Maximum and Minimum Temperatures')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')

# Add legend
plt.legend()

# Display the chart
plt.show()
```

```python
# Convert the 'Date' Column to datetime format
weather_clean['date'] = pd.to_datetime(weather_clean['date'], format='%Y%m%d') # Adjust the data format to date format
weather_clean.head()
```
